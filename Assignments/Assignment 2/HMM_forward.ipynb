```python
import numpy as np
```

```python
# f_1:t+1 = alpha * O_t+1 T_transposed f_1:t
O_T = np.array([[0.9, 0] , [0, 0.2]])
O_F = np.array([[0.1, 0] , [0, 0.8]])


T = np.array([[0.7, 0.3], [0.3, 0.7]])
T_transposed = np.transpose(T)

def forward(U):
    f = np.zeros((len(U)+1, 2))     # f is indexed 1 timestep ahead of U
    f[0] = np.array([0.5, 0.5])
    for i in range(len(U)):
        O_i = O_T if U[i] else O_F
        unnormalized_f = O_i @ T_transposed @ f[i]
        normalized_f = unnormalized_f / np.sum(unnormalized_f)
        f[i+1] = normalized_f
    return f
```

```python
U = [True, True]

forward(U)

U = [True, True, False, True, True]

forward(U)
```
